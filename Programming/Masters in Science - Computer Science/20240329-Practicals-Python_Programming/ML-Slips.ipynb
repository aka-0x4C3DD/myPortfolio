#### Q1) Load the Wine dataset from scikit-learn. Preprocess the data: Check for missing values and handle them appropriately (e.g., impute with mean/median or remove rows). Consider normalization or standardization of features if the scales differ significantly (e.g., using MinMaxScaler or StandardScaler from scikit-learn). Implement a KNN model to predict the type of wine (class labels) for new data points using the KNeighborsClassifier from scikit-learn. Evaluate the model’s accuracy on a split testing set.

##### Load the Wine dataset from scikit-learn:

```python
from sklearn.datasets import load_wine

# Load the dataset
wine_data = load_wine()
X = wine_data.data
y = wine_data.target
```

##### Preprocess the data: Check for missing values and handle them appropriately (e.g., impute with mean/median or remove rows):

```python
import numpy as np
from sklearn.impute import SimpleImputer

# Check for missing values
missing_values = np.isnan(X)

# If there are missing values, impute them
if np.any(missing_values):
    imputer = SimpleImputer(strategy='mean')
    X = imputer.fit_transform(X)
```

##### Consider normalization or standardization of features if the scales differ significantly:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

##### Implement a KNN model to predict the type of wine (class labels) for new data points:

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Create and train the KNN model
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

##### Evaluate the model’s accuracy on a split testing set:

```python
from sklearn.metrics import accuracy_score

# Make predictions on the testing set
y_pred = knn_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

#### Q5) Load the Wine dataset from scikit-learn. Preprocess the data: Check for missing values. Implement K-Means clustering to segment the iris data into 3 clusters. Visualize the clusters using a scatter plot, coloring data points based on their assigned cluster labels (e.g., using matplotlib).

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

# Load the Wine dataset
wine_data = load_wine()
X = wine_data.data

# Check for missing values
missing_values = np.isnan(X).sum()
print("Missing values:", missing_values)

# Since there are no missing values, we can proceed with K-Means clustering

# Implement K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)

# Visualize the clusters using a scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot for the first two features
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', edgecolor='k', s=50)
plt.xlabel(wine_data.feature_names[0])
plt.ylabel(wine_data.feature_names[1])
plt.title("K-Means Clustering of Wine Dataset (First two features)")

plt.colorbar(label='Cluster')

plt.show()
```

#### Q6) Load the Iris dataset from scikit-learn. Preprocess the data: Check for missing values. Implement K-Means clustering to segment the iris data into 3 clusters. Visualize the clusters using a scatter plot, coloring data points based on their assigned cluster labels (e.g., using matplotlib).

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_data = load_iris()
X = iris_data.data

# Check for missing values (Iris dataset is clean, so we skip this step)

# Implement K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster labels
cluster_labels = kmeans.labels_

# Visualize the clusters using a scatter plot
plt.figure(figsize=(8, 6))

# Plot the data points colored by cluster labels
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-Means Clustering of Iris Dataset')
plt.colorbar(label='Cluster Label')
plt.show()
```
