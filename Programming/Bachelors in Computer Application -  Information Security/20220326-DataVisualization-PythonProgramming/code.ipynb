```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.legend_handler as hl2d
```

```python
df = pd.read_excel(r'SYB64_314_202110_Internet Usage.xlsx')
df
```

```python
total = df[df.Areas == 'Total, all countries or areas']
naf = df[df.Areas == 'Northern Africa']
ssaf = df[df.Areas == 'Sub-Saharan Africa']
eaf = df[df.Areas == 'Eastern Africa']
maf = df[df.Areas == 'Middle Africa']
saf = df[df.Areas == 'Southern Africa']
waf = df[df.Areas == 'Western Africa']
nam = df[df.Areas == 'Northern America']
car = df[df.Areas == 'Caribbean']
lam = df[df.Areas == 'Latin America']
cas = df[df.Areas == 'Central Asia']
eas = df[df.Areas == 'Eastern Asia']
scas = df[df.Areas == 'South-central Asia']
seas = df[df.Areas == 'South-eastern Asia']
sas = df[df.Areas == 'Southern Asia']
wa = df[df.Areas == 'Western Asia']
eu = df[df.Areas == 'Europe']
oc = df[df.Areas == 'Oceania']
anz = df[df.Areas == 'Australia and New Zealand']
anz
```

```python
plt_1 = plt.figure(figsize = (20,10))

line1, = plt.plot(total.Year, total.Value, marker = '$T$', label = 'Total')
line2, = plt.plot(naf.Year, naf.Value, marker = '*', label = 'Northern Africa')
line3, = plt.plot(ssaf.Year, ssaf.Value, marker = 'D', label = 'Sub-Saharan Africa')
line4, = plt.plot(eaf.Year, eaf.Value, marker = '^', label = 'Eastern Africa')
line5, = plt.plot(maf.Year, maf.Value, marker = '2', label = 'Middle Africa')
line6, = plt.plot(saf.Year, saf.Value, marker = 's', label = 'Southern Africa')
line7, = plt.plot(waf.Year, waf.Value, marker = 'h', label = 'Western Africa')
line8, = plt.plot(nam.Year, nam.Value, marker = 'o', label = 'Northern America')
line9, = plt.plot(car.Year, car.Value, marker = 'X', label = 'Caribbean')
line10, = plt.plot(lam.Year, lam.Value, marker = 'P', label = 'Latin America')
line11, = plt.plot(cas.Year, cas.Value, marker = 'p', label = 'Central Asia')
line12, = plt.plot(eas.Year, eas.Value, marker = '|', label = 'Eastern Asia')
line13, = plt.plot(scas.Year, scas.Value, marker = '_', label = 'South-central Asia')
line14, = plt.plot(seas.Year, seas.Value, marker = 'v', label = 'South-eastern Asia')
line15, = plt.plot(sas.Year, sas.Value, marker = '1', label = 'Southern Asia')
line16, = plt.plot(wa.Year, wa.Value, marker = 'x', label = 'Western Asia')
line17, = plt.plot(eu.Year, eu.Value, marker = '+', label = 'Europe')
line18, = plt.plot(oc.Year, oc.Value, marker = 'd', label = 'Oceania')
line19, = plt.plot(anz.Year, anz.Value, marker  = '>', label = 'Australia and New Zealand')
plt.title("Internet Usage through out the World in Two Decades", fontsize = 32, pad = 1)
plt.legend(bbox_to_anchor=(1.01, 0.11), loc=3, borderaxespad=0., handles= [line1, line2, line3, line4, line5, line6, line7, line8, line9, line10, line11, line12, line13, line14, line15, line16, line17, line18, line19], fontsize = 15)
plt.xlabel('Years ------>', fontsize = 20)
plt.ylabel('Values in % ------>', fontsize = 20)
plt.show()
```
